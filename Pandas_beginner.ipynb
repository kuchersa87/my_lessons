## 2.7 Основы Numpy. Маска и слайсинг вместе, прихотливая индексация, а еще reshape

```python
# В скобках можно работать и с маской и с индексами одновременно!
import numpy as np
users = np.array(['Andry', 'Roy', 'Jim', 'Carl', 'Tomas']) # имя посетителей
# Каждая строка соответствует имени посетителя
visits = np.array([
    [1,2,1,0,4,0,2],  # день недели / количество посещений в каждый день
    [1,3,1,2,0,1,3],
    [4,1,5,0,3,6,2],
    [1,2,1,3,0,1,3],
    [0,2,0,0,5,2,2],
])
mask = users == 'Roy'
print(f'Фильтр через mask даст нам {visits[mask]}')  # Вытащит данные за всю неделю
# Теперь нам необходиимы вытащить данные только за субботу и воскресенье
mask2 = users == 'Jim'
print(visits[mask2])  # Пробуем через другую маску -Jim
print(visits[mask2, 5:])  # Маска + срез(слайс) - суббота и восресенье
print(visits[mask2, -1])  # воскресенье
```

```python
# Что делать, если не Адрей?
mask3 = users != 'Tomas'  # все, кроме томаса
print(mask3)
print(visits[mask3])
```

```python
# NOT в Numpy
# Получить противоположную маску
# Специальная команда в Numpy - позволяет инверсировать - знак тильда ~
mask4 = users == 'Carl'
print(mask4)   # Обычная маска
print(~mask4)  # инверсированная версия маски
```

```python
# При работе с numpy для логических операций вы не работаете со стандартными питоновскими and и or.
# У numpy свои операторы:
# Логическое ИЛИ: |
# Логическое И: &
# OR +
# AND *
# False 0
# True 1
# True or False (1+0=1) True
# True and False (1*0=0) False
# Объединение масок:
print("   ", mask4)  # готовая маска
print("   ", mask2)  # готовая маска
print("И  ", mask2 & mask4)  # условие И
print("ИЛИ", mask2 | mask4)  # условие ИЛИ
mask5 = (users == 'Jim') & (users == 'Carl')  # двойная маска с условием И
mask6 = (users == 'Jim') | (users == 'Carl')  # двойная маска с условием ИЛИ
print(mask5)
print(mask6)
#mask5 = (mask2) & (mask4)
#mask6 = (mask2) | (mask4)
print("Массив через общую маску 6:", visits[mask6])
print("Массив через общую маску 5:", visits[mask5])
```

```python
# Использование параметра how в функции dropna
# Параметром how мы можем регулировать условие при котором функция dropna будет выполнять удаление строк.
# how ='all' - в этом случае, если во всей строке  пропуски, тогда строка удаляется
# how ='any' - в этом случае, если в строке есть хотя бы один пропуск, то строка удаляется.
# По умолчанию у dropna параметр howравен 'any'
```

```python
# Удалим строки, где все строки с пропусками
df1.dropna(how='all')
```

```python
# Удалим строки, где имеется хотябы 1 пропуск
df1.dropna(how='any')
```

```python
# Пример посложнее, где несколько переменных в одной ячейке
# По умолчаю функция не сможет корректно разбить колонки
films = pd.DataFrame([['Анчартед', 115, 'боевик|приключения'], 
                      ['Доктор Стрэндж', 126, 'фантастика|боевик|приключения'], 
                      ['Соник', 122, 'боевик|приключения'], 
                      ['Одиннадцать', 121, 'драма|спорт'], 
                      ['Тор', 119, 'фантастика|боевик'], 
                      ['Мистер нокаут', 118, 'спорт'],
                      ['Я краснею', 100, 'приключения']], columns=['Film', 'Duration', 'Genre'])
# Вытащим уникальные значения жанров через цикл
all_genres = []
for x in films.Genre:
  all_genres.extend(x.split('|'))   # разбили по знаку и добавили все значения в список all_genres
genres = pd.unique(all_genres)      # взял только уникальные значения
# Создадим таблицу с 0, которую заполним соответстенно
m = np.zeros((len(films), len(genres)), dtype='int')
# Создаём матрицу, которую будем клеить к дф
matrix = pd.DataFrame(m, columns = genres)
# Составим цикл, который будет сам подставлять 1 там, где это необходимо
for i, genre in enumerate(films.Genre):
  idx = matrix.columns.get_indexer(genre.split('|'))
  matrix.iloc[i, idx] = 1
# Приклеим к основному дф
films = films[['Film', 'Duration']].join(matrix.add_prefix('genre_'))
films
```

```python
# Данную функцию также можно использовать для количественных переменных
# Разделить фильмы на 2 группы по продолжительности фильма
# Для это функцию dummies необходимо использовать совместно с функциями cut / qcut

films = pd.DataFrame([['Анчартед', 115, 'боевик|приключения'], 
                      ['Доктор Стрэндж', 126, 'фантастика|боевик|приключения'], 
                      ['Соник', 122, 'боевик|приключения'], 
                      ['Одиннадцать', 121, 'драма|спорт'], 
                      ['Тор', 119, 'фантастика|боевик'], 
                      ['Мистер нокаут', 118, 'спорт'],
                      ['Я краснею', 100, 'приключения']], columns=['Film', 'Duration', 'Genre'])
# Создадим матрицу с разделением
matrix = pd.get_dummies(pd.qcut(films.Duration, 2))  # 2 части / колонки
films = films.join(matrix.add_prefix('duration_'))
films
```
