## https://stepik.org/lesson/667236/step/9?unit=665239

```python
import pandas as pd
import numpy as np
def solution():
    s1 = pd.Series([1, 2, 3, 4, 5])
    s2 = pd.Series([4, 5, 6, 7, 8])
    return s1[~s1.isin(s2)]
   # s1[s1.isin(s2) == False]
solution()
```
