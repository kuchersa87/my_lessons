```python
2.1 –ú–µ—Ç–æ–¥—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞. –ê—Ä–≥—É–º–µ–Ω—Ç self

# –ú–µ—Ç–æ–¥ - —ç—Ç–∞ —Ç–∞ –∂–∞ —Ñ—É–Ω–∫—Ü–∏—è, –Ω–æ –æ–±—ä—è–≤–ª–µ–Ω–Ω–∞—è –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Å–∞
# –ú–µ—Ç–æ–¥ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Ñ—É–Ω–∫—Ü–∏–∏ —Ç–µ–º, —á—Ç–æ –º–µ—Ç–æ–¥ –±—É–¥–µ—Ç –ø—Ä–∏–≤—è–∑–∞–Ω –∫ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º—É –æ–±—ä–µ–∫—Ç—É - –≤—ã–∑–≤–∞—Ç—å –º–æ–∂–Ω–æ –≤–æ—Ç —Ç–∞–∫: bob.hello()
# –ü—Ä–∏ –≤—ã–∑–æ–≤–µ –º–µ—Ç–æ–¥–∞ —Ç–æ—Ç –æ–±—ä–µ–∫—Ç, —Å –∫–æ—Ç–æ—Ä—ã–º –æ–Ω —Å–≤—è–∑–∞–Ω –±—É–¥–µ—Ç –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–æ—Å—Ç–∞–≤–ª—è—Ç—å—Å—è –≤ –∞—Ä–≥—É–º–µ–Ω—Ç —Ñ—É–Ω–∫—Ü–∏–∏
‚Äã
# –ú–µ—Ç–æ–¥ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å–≤—è–∑—ã–≤–∞–µ—Ç—Å—è —Å –æ–±—ä–µ–∫—Ç–æ–º, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –æ–Ω –±—ã–ª –≤—ã–∑–≤–∞–Ω - bob.hello()
# self - –æ–±—â–µ–ø—Ä–∏–Ω—è—Ç–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–∞, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –±—ã–ª –≤—ã–∑–≤–∞–Ω –º–µ—Ç–æ–¥
# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π –∫–ª–∞—Å—Å –∏ —Ñ—É–Ω–∫—Ü–∏—é –≤–Ω—É—Ç—Ä–∏ –Ω–µ–≥–æ, –∫–∞–∫ –∞—Ç—Ä–∏–±—É—Ç
class Cat:
    def hello():                # –≠—Ç–æ –º–µ—Ç–æ–¥ / —Ñ—É–Ω–∫—Ü–∏—è
        print('Hello kitty!')   # 
# –û–±—Ä–∞—â–µ–Ω–∏–µ –∫ —Ñ—É–Ω–∫—Ü–∏–∏ —á–µ—Ä–µ–∑ –∫–ª–∞—Å—Å
print(Cat.hello)
print(Cat.hello())
# <function Cat.hello at 0x0000023B9DF6E670>
# Helli kitty!

# –°–æ–∑–¥–∞–¥–∏–º –≠–ö
bob = Cat()
# –û–±—Ä–∞—Ç–∏–º—Å—è –∫ —Ñ—É–Ω–∫—Ü–∏–∏ —á–µ—Ä–µ–∑ –≠–ö

bob.hello()
# TypeError: hello() takes 0 positional arguments but 1 was given
# –ó–∞–ø–æ–ª–Ω–∏–ª–∏ –ø–æ–ª–µ —Å –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏
class Cat:
    def hello(*args):  # –¥–æ–±–∞–≤–∏–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∞—Ä–≥—É–º–µ–Ω—Ç—ã –¥–ª—è –≤–æ–∑–º–æ–∂–Ω–æ—Ç—Å—Ç–∏ –≤—ã–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü–∏–∏
        print('Hello kitty!', args)
jim = Cat()
jim.hello()
class Cat:
    breed = 'pers'
    def hello(*args):                  # —Ñ—É–Ω–∫—Ü–∏—è –ø–µ—á–∞—Ç–∏
        return 'Hello kitty!', args
    def show_breed(instance):          # —Ñ—É–Ω–∫—Ü–∏—è —É–∫–∞–∑–∞–Ω–∏—è –ø–æ–ø—Ä–æ–¥—ã
        return f'my breed is {instance.breed}'
    def show_name(instance):           # —Ñ—É–Ω–∫—Ü–∏—è —É–∫–∞–∑–∞–Ω–∏—è –∏–º–µ–Ω–∏
        if hasattr(instance, 'name'):  # –µ—Å–ª–∏ –µ—Å—Ç—å –∏–º—è - —É–∫–∞–∑—ã–≤–∞–µ–º –µ–≥–æ
            return f'my name is {instance.name}'
        else:                          # –µ—Å–ª–∏ –Ω–µ—Ç –∏–º–µ–Ω–∏ - –≤—ã–≤–æ–¥–∏–º –Ω–∞–¥–ø–∏—Å—å –æ–± —ç—Ç–æ–º
            return 'no name'
    def set_name(koshka, value):       # —Ñ—É–Ω–∫—Ü–∏—è –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–∞ –∏–º–µ–Ω–∏
        koshka.name = value
    def set_show_age(koshka, value):   # —Ñ—É–Ω–∫—Ü–∏—è –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–∞ –≤–æ–∑—Ä–∞—Å—Ç–∞ –∏ –µ–≥–æ –≤—ã–≤–æ–¥–∞ –Ω–∞ –ø–µ—á–∞—Ç—å
        koshka.age = value
        return f'I am {koshka.age} years old'
    def set_show_fish(self, fish=0):   # –§—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞–∑–Ω–∞—á–∏—Ç –∏ –≤—ã–≤–µ–¥–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ã–±—ã —É –∫–æ—Ç–∞
        self.fish = fish
        return f'I have {self.fish} fish'
walt = Cat()
print(walt.hello())
print('walt said', walt.show_breed())
walt.breed = 'siam'
print('walt said', walt.show_breed())
mary = Cat()
mary.name = 'MARY'
print('mary said', mary.show_name())
print('walt said', walt.show_name())
tom = Cat()
tom.set_name('TOM')
print(tom.show_name())
print('TOM said', tom.set_show_age(90))
print(tom.set_show_fish())

–ó–∞–¥–∞—á–∞ 1 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Lion. –í –Ω–µ–º –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –º–µ—Ç–æ–¥ roar, –∫–æ—Ç–æ—Ä—ã–π –ø–µ—á–∞—Ç–∞–µ—Ç –Ω–∞ —ç–∫—Ä–∞–Ω "Rrrrrrr!!!"
# –ù–∞–ø–∏—à–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ Lion       
class Lion:
    def roar(self):
        print('Rrrrrrr!!!')
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–ª–∞—Å—Å–∞ Lion        
s = Lion()
assert isinstance(s, Lion)
assert s.__dict__ == {}
s.roar()

–ó–∞–¥–∞—á–∞ 2 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Robot, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω—ã: –ú–µ—Ç–æ–¥ say_hello , –ø–µ—á–∞—Ç–∞—é—â–∏–π –Ω–∞ —ç–∫—Ä–∞–Ω —Ñ—Ä–∞–∑—É ¬´Hello, human! My name is C-3PO¬ª –ú–µ—Ç–æ–¥ say_bye , –ø–µ—á–∞—Ç–∞—é—â–∏–π –Ω–∞ —ç–∫—Ä–∞–Ω —Ñ—Ä–∞–∑—É ¬´See u later alligator¬ª –ü–æ—Å–ª–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ —Å–æ–∑–¥–∞–π—Ç–µ 2 —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –∏—Ö –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ c3po –∏ r2d2 –ó–∞—Ç–µ–º –≤—ã–∑–æ–≤–∏—Ç–µ —É –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π c3po –º–µ—Ç–æ–¥ say_hello , –∞ –∑–∞—Ç–µ–º –º–µ—Ç–æ–¥ say_bye –ò —Ç–æ –∂–µ —Å–∞–º–æ–µ —Å–¥–µ–ª–∞–π—Ç–µ —Å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π r2d2: –≤—ã–∑–æ–≤–∏—Ç–µ –º–µ—Ç–æ–¥ say_hello , –ø–æ—Ç–æ–º ‚Äî –º–µ—Ç–æ–¥ say_bye
class Robot:
    def say_hello(self):
        print('Hello, human! My name is C-3PO')
    def say_bye(self):
        print('See u later alligator')
c3po = Robot()
r2d2 = Robot()
c3po.say_hello()
c3po.say_bye()
r2d2.say_hello()
r2d2.say_bye()

–ó–∞–¥–∞—á–∞ 3
–í –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–∞–¥–∞—á–µ –≤—ã –º–æ–≥–ª–∏ –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç–æ, —á—Ç–æ –≤—ã–≤–æ–¥–∏—Ç—Å—è –≤—Å–µ–≥–¥–∞ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –∏–º—è —Ä–æ–±–æ—Ç–∞ –≤ –º–µ—Ç–æ–¥–µ say_hello . 
–î–∞–≤–∞–π—Ç–µ —ç—Ç–æ –∏—Å–ø—Ä–∞–≤–∏–º –ø—Ä–∏ –ø–æ–º–æ—â–∏ –∞—Ç—Ä–∏–±—É—Ç–∞ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ name . –î–ª—è —ç—Ç–æ–≥–æ –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–ª–∞—Å—Å Robot, –≤ –∫–æ—Ç–æ—Ä–æ–º –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å 
—Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω—ã:
–ú–µ—Ç–æ–¥ set_name , –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π –∏–º—è —Ä–æ–±–æ—Ç–∞ –∏ —Å–æ—Ö—Ä–∞–Ω—è—é—â–∏–π –µ–≥–æ –≤ –∞—Ç—Ä–∏–±—É—Ç–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ name 
–ú–µ—Ç–æ–¥ say_hello . –ú–µ—Ç–æ–¥ –¥–æ–ª–∂–µ–Ω –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, –µ—Å—Ç—å –ª–∏ —É –≠–ö –∞—Ç—Ä–∏–±—É—Ç name . –ï—Å–ª–∏ –∞—Ç—Ä–∏–±—É—Ç name  –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å 
—Ñ—Ä–∞–∑—É ¬´Hello, human! My name is <name>¬ª. –ï—Å–ª–∏ –∞—Ç—Ä–∏–±—É—Ç name  –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç —É —ç–∫–∑–µ–º–ø–ª—è—Ä–∞, –ø–µ—á–∞—Ç–∞–π—Ç–µ —Å–æ–æ–±—â–µ–Ω–∏–µ ¬´–£ —Ä–æ–±–æ—Ç–∞ –Ω–µ—Ç –∏–º–µ–Ω–∏¬ª 
–ú–µ—Ç–æ–¥ say_bye ,  –ø–µ—á–∞—Ç–∞—é—â–∏–π –Ω–∞ —ç–∫—Ä–∞–Ω —Ñ—Ä–∞–∑—É ¬´See u later alligator¬ª
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ Robot
üöÄ–ü–æ–¥—Å–∫–∞–∑–∫–∞üöÄ
–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞–ª–∏—á–∏—è –∞—Ç—Ä–∏–±—É—Ç–∞ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ—É–Ω–∫—Ü–∏–µ–π hasattr
# –ù–∞–ø–∏—à–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ Robot       
class Robot:
    def set_name(self, name):
        self.name = name
    def say_hello(self):
        if hasattr(self, 'name'):
            print(f'Hello, human! My name is {self.name}')
        else:
            print('–£ —Ä–æ–±–æ—Ç–∞ –Ω–µ—Ç –∏–º–µ–Ω–∏')
    def say_bye(self):
        print('See u later alligator')
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–ª–∞—Å—Å–∞ Robot 
c3po = Robot()
c3po.say_hello()
c3po.set_name('R2D2')
c3po.say_hello()
c3po.say_bye()
‚Äã
r = Robot()
r.set_name('Chappy')
r.say_hello()
‚Äã
d = Robot()
d.say_hello()
d.set_name('Wally')
d.say_hello()

–ó–∞–¥–∞—á–∞ 4 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Counter, —ç–∫–∑–µ–º–ø–ª—è—Ä—ã –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥—É—Ç –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞—Ç—å –≤–Ω—É—Ç—Ä–∏ —Å–µ–±—è –∑–Ω–∞—á–µ–Ω–∏—è. –í –∫–ª–∞—Å—Å–µ Counter –Ω—É–∂–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å: –ú–µ—Ç–æ–¥ start_from, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç. –ó–Ω–∞—á–µ–Ω–∏–µ, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –ø–æ–¥—Å—á–µ—Ç, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é —Ä–∞–≤–Ω–æ 0. –ú–µ—Ç–æ–¥ increment, –∫–æ—Ç–æ—Ä—ã–π —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç —Å—á–µ—Ç—á–∏–∫ –Ω–∞ 1. –ú–µ—Ç–æ–¥ display, –∫–æ—Ç–æ—Ä—ã–π –ø–µ—á–∞—Ç–∞–µ—Ç —Ñ—Ä–∞–∑—É "–¢–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—á–µ—Ç—á–∏–∫–∞ = ". –ú–µ—Ç–æ–¥ reset, –∫–æ—Ç–æ—Ä—ã–π –æ–±–Ω—É–ª—è–µ—Ç –Ω–∞–∫–æ–ø–∏–≤—à–µ–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ —Å—á–µ—Ç—á–∏–∫–∞.

class Counter:
    def start_from(self, start = 0):
        self.start = start
    def increment(self):
        self.start += 1
    def display(self):
        print(f'–¢–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—á–µ—Ç—á–∏–∫–∞ = {self.start}')
    def reset(self):
        self.start = 0
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–ª–∞—Å—Å–∞ Counter 
c1 = Counter()
c2 = Counter()
assert isinstance(c1, Counter)
assert isinstance(c2, Counter)
assert c1.__dict__ == {}
assert c2.__dict__ == {}
c1.start_from(45)
assert len(c1.__dict__) == 1
c1.increment()
c1.display()  # –ø–µ—á–∞—Ç–∞–µ—Ç 46
c1.increment()
c1.display()  # –ø–µ—á–∞—Ç–∞–µ—Ç 47
c2.start_from()
c2.display()  # –ø–µ—á–∞—Ç–∞–µ—Ç 0
c2.increment()
c2.display()  # –ø–µ—á–∞—Ç–∞–µ—Ç 1
c1.reset()  # –æ–±–Ω—É–ª–∏–ª–∏ —Å1, –Ω–æ —Å2 –Ω–µ –¥–æ–ª–∂–µ–Ω –º–µ–Ω—è—Ç—å—Å—è
c1.display()  # –ø–µ—á–∞—Ç–∞–µ—Ç 0
c2.display()  # –ø–æ–ø—Ä–µ–∂–Ω–µ–º—É –ø–µ—á–∞—Ç–∞–µ—Ç 1

–ó–∞–¥–∞—á–∞ 5 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Constructor, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω—ã: –ú–µ—Ç–æ–¥ add_atribute , –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π –Ω–∞ –≤—Ö–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–∞ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏ –∏ –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ. –ü—Ä–∏ –ø–æ–º–æ—â–∏ —Ñ—É–Ω–∫—Ü–∏–∏ setattr –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ–∑–¥–∞—Ç—å –∏–ª–∏ –∏–∑–º–µ–Ω–∏—Ç—å –∞—Ç—Ä–∏–±—É—Ç –¥–ª—è –≠–ö, —É –∫–æ—Ç–æ—Ä–æ–≥–æ —ç—Ç–æ—Ç –º–µ—Ç–æ–¥ –±—ã–ª –≤—ã–∑–≤–∞–Ω. –ú–µ—Ç–æ–¥ display , –ø–µ—á–∞—Ç–∞—é—â–∏–π –Ω–∞ —ç–∫—Ä–∞–Ω —Å–ª–æ–≤–∞—Ä—å dict —É –≠–ö. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ Constructor

# –ù–∞–ø–∏—à–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ Constructor       
class Constructor:
    def add_atribute(self, atr, value):
        setattr(self, atr, value)
    def display(self):
        print(self.__dict__)
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–ª–∞—Å—Å–∞ Constructor 
obj1 = Constructor()
assert obj1.__dict__ == {}
obj1.display()
obj1.add_atribute('color', 'red')
assert obj1.color == 'red', 'wrong color'
obj1.add_atribute('width', 20)
assert obj1.width == 20
obj1.display()

–ó–∞–¥–∞—á–∞ 6 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Point. –£ —ç—Ç–æ–≥–æ –∫–ª–∞—Å—Å–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å: –ú–µ—Ç–æ–¥ set_coordinates, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç–æ—á–∫–∏ –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –∏—Ö –≤ —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ –≤ –∞—Ç—Ä–∏–±—É—Ç—ã x –∏ y –ú–µ—Ç–æ–¥ get_distance, –∫–æ—Ç–æ—Ä—ã–π –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ Point –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏ –ø–æ —Ç–µ–æ—Ä–µ–º–µ –ü–∏—Ñ–∞–≥–æ—Ä–∞. –í —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ –≤ –¥–∞–Ω–Ω—ã–π –º–µ—Ç–æ–¥ –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –Ω–µ —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ Point, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ —Å–æ–æ–±—â–µ–Ω–∏–µ "–ü–µ—Ä–µ–¥–∞–Ω–∞ –Ω–µ —Ç–æ—á–∫–∞".

class Point:
    def set_coordinates(self, point1, point2):
        setattr(self, 'x', point1)
        setattr(self, 'y', point2)
    def get_distance(self, point):
        if isinstance(point, Point):
            return ((point.x - self.x)**2 + (point.y - self.y)**2)**0.5
        else:
            print('–ü–µ—Ä–µ–¥–∞–Ω–∞ –Ω–µ —Ç–æ—á–∫–∞')
            return None
# –ö–æ–¥ –Ω–∏–∂–µ –Ω–µ —É–¥–∞–ª—è–π—Ç–µ, –æ–Ω –Ω—É–∂–µ–Ω –¥–ª—è –ø—Ä–æ–≤–µ—Ä–æ–∫
p1 = Point()
p2 = Point()
assert isinstance(p1, Point)
assert isinstance(p2, Point)
‚Äã
p1.set_coordinates(1, 2)
assert p1.x == 1
assert p1.y == 2
p2.set_coordinates(4, 6)
assert p2.x == 4
assert p2.y == 6
assert p1.get_distance(p2) == 5.0
p3 = Point()
p3.set_coordinates(10, 10)
p1.set_coordinates(4, 2)
assert p1.get_distance(p3) == 10.0
res = p1.get_distance(10)  # –†–∞—Å–ø–µ—á–∞—Ç–∞–µ—Ç "–ü–µ—Ä–µ–¥–∞–Ω–∞ –Ω–µ —Ç–æ—á–∫–∞", –≤–µ—Ä–Ω–µ—Ç None
assert res is None, '–ú–µ—Ç–æ–¥ get_distance –¥–æ–ª–∂–µ–Ω –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å None, –µ—Å–ª–∏ –≤ –Ω–µ–≥–æ –±—ã–ª–∞ –ø–µ—Ä–µ–¥–∞–Ω–∞ –Ω–µ —Ç–æ—á–∫–∞'
‚Äã
assert p2.get_distance([1, 2, 3]) is None  # –†–∞—Å–ø–µ—á–∞—Ç–∞–µ—Ç "–ü–µ—Ä–µ–¥–∞–Ω–∞ –Ω–µ —Ç–æ—á–∫–∞", –≤–µ—Ä–Ω–µ—Ç None
```

```python
2.2 –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –æ–±—ä–µ–∫—Ç–∞. –ú–µ—Ç–æ–¥ init

# –ö–∞–∂–¥—ã–π —Ä–∞–∑ –≤—ã–∑—ã–≤–∞—Ç—å –º–µ—Ç–æ–¥ –¥–ª—è –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –¥–ª—è –≠–ö –Ω–µ—É–¥–æ–±–Ω–æ:
class Robot:
    def set_name(self, name):
        self.name = name
bob = Robot()
bob.set_name('BOB')
bob.name
# –ü–µ—Ä–≤—ã–π –º–∞–≥–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ init –Ω—É–∂–µ–Ω –¥–ª—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ / –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –Ω–∞—à–µ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∫–∞–∫–∏–º–∏-–ª–∏–±–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
# –í–∑–∞–º–µ–Ω –≤—ã–∑–æ–≤–∞ –º–µ—Ç–æ–¥–∞ setattr –∫–∞–∂–¥—ã–π —Ä–∞–∑
# –ï—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º –ø—Ä–æ—Å—Ç–∞–≤–ª—è—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –≠–ö, —Ç–æ –¥–ª—è —ç—Ç–æ–≥–æ –≤ Python –µ—Å—Ç—å –º–µ—Ç–æ–¥ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ __init__
# –ú–∞–≥–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ __init__ –±—É–¥–µ—Ç —Å—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –æ–±—ä–µ–∫—Ç–∞. 
# –¶–µ–ª—å –º–µ—Ç–æ–¥–∞ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ —Å–æ—Å—Ç–æ–∏—Ç –≤ —Ç–æ–º, —á—Ç–æ–±—ã –ø—Ä–æ—Å—Ç–∞–≤–∏—Ç—å –Ω–∞—à –≤–Ω–æ–≤—å —Å–æ–∑–¥–∞–Ω–Ω—ã–π –≠–ö –Ω—É–∂–Ω—ã–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏.
# –ú–µ—Ç–æ–¥ __init__ –≤—Å–µ–≥–¥–∞ –¥–æ–ª–∂–µ–Ω –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å None
‚Äã
class Cat:
    def __init__(self, name, age = 40, color = 'blue'):  # —É–∫–∞–∑—ã–≤–∞–µ–º –ø—Ä–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –≠–ö. age –∏ color —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
        print('method init')  # –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –æ—Ç—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω–æ–≥–æ –±–ª–æ–∫–∞
        self.name, self.age, self.color = name, age, color  # –∑–∞–¥–∞—ë–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
tom = Cat('Tom', 10)  # –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–µ—Ä–µ–¥–∞—Ç—å –æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –Ω–∞–∑–Ω–∞—á–∞—Ç—Å—è
tom.__dict__
# {'name': 'Tom', 'age': 10, 'color': 'blue'}
# –ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –≠–ö Python —Å–Ω–∞—á–∞–ª–∞ –∑–∞–ø—É—Å–∫–∞–µ—Ç –º–µ—Ç–æ–¥-–∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä __new__ –¥–ª—è –Ω–æ–≤–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∏ –ø–æ —Å—É—Ç–∏ —Å–æ–∑–¥–∞—ë—Ç 
# –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –∏–º—ë–Ω (–∞—Ç—Ä–∏–±—É—Ç __dict__), –ø–æ—Å–ª–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –∑–∞–ø—É—Å–∫–∞–µ—Ç—Å—è –º–µ—Ç–æ–¥ __init__ , –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É—é—â–∏–π –∞—Ç—Ä–∏–±—É—Ç—ã.
# –ü–æ—Å–∫–æ–ª—å–∫—É –¥–∞–Ω–Ω—ã–µ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å–ª–æ–≤–∞—Ä–µ –∏ –≤—ã–∑—ã–≤–∞—é—Ç—Å—è –º–µ—Ç–æ–¥–æ–º __dict__, 
# –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥—ã —Å–ª–æ–≤–∞—Ä–µ–π –¥–ª—è –≤—ã–≤–æ–¥–∞ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –∏–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –∫–ª–∞—Å—Å–∞:
class Car:
    "–ö–ª–∞—Å—Å –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –º–∞—à–∏–Ω"
    def __init__(self, new_model='Car', new_engine=None, new_horse_power=0):
        print('method init')
        self.model = new_model   
        self.engine = new_engine
        self.horse_power = new_horse_power
auto = Car('BMW', 2.5, 350)
print(auto.__dict__)            # –ø—Ä–æ—Å—Ç–æ –≤—ã–≤–æ–¥ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –≤ —Å–ª–æ–≤–∞—Ä–µ
print(*auto.__dict__.values())  # –ø–æ–ª—É—á–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç–æ–≤
print(*auto.__dict__.items())   # –ø–æ–ª—É—á–µ–Ω–∏–µ —Ä–∞—Å–ø–∞–∫–æ–≤–∞–Ω–Ω—ã—Ö –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –≤ —Ñ–æ—Ä–º–µ –∫–æ—Ä—Ç–µ–∂–∞
print(*auto.__dict__.keys())    # –ø–æ–ª—É—á–µ–Ω–∏–µ –∫–ª—é—á–µ–π —Å–ª–æ–≤–∞—Ä—è
–ó–∞–¥–∞—á–∞ 1 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Vehicle, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –µ—Å—Ç—å: –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä init, –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é —Å–∫–æ—Ä–æ—Å—Ç—å –∏ –ø—Ä–æ–±–µ–≥. –ò—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∞—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ max_speed –∏ mileage —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed, self.mileage = max_speed, mileage
# –ù–∏–∂–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–æ–∫, –µ–≥–æ –Ω–µ –Ω—É–∂–Ω–æ —É–¥–∞–ª—è—Ç—å        
modelX = Vehicle(200, 18000)
assert modelX.max_speed == 200
assert modelX.mileage == 18000
assert modelX.__dict__ == {'max_speed': 200, 'mileage': 18000}
audi = Vehicle(240, 5)
assert audi.__dict__ == {'max_speed': 240, 'mileage': 5}
print('Good')
–ó–∞–¥–∞—á–∞ 2 –ü–µ—Ä–µ–¥ –í–∞—Å–∏–ª–∏–µ–º –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –∑–∞–¥–∞—á—É —Å–æ–∑–¥–∞—Ç—å –∫–ª–∞—Å—Å Person –∏ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –≤ –Ω–µ–º —Å–ª–µ–¥—É—é—â–∏–µ –º–µ—Ç–æ–¥—ã: init –º–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–æ–≤ name –∏ age –º–µ—Ç–æ–¥ greet, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–π —Å—Ç—Ä–æ–∫—É –≤ —Å–ª–µ–¥—É—é—â–µ–º —Ñ–æ—Ä–º–∞—Ç–µ: ¬´Hello, my name is [name], and I am [age] years old¬ª –í–æ—Ç —á—Ç–æ –Ω–∞—à–∫–æ–¥–∏–ª –í–∞—Å–∏–ª–∏–π, –ø–æ—Å–º–æ—Ç—Ä–∏—Ç–µ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –∫–æ–¥–∞ –Ω–∏–∂–µ. –û–Ω –Ω–µ –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø—Ä–æ–≤–µ—Ä–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–ø–∏—Å–∞–Ω—ã –Ω–∏–∂–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ Person. –í–∞—à–∞ –∑–∞–¥–∞—á–∞ –Ω–∞–π—Ç–∏ –æ—à–∏–±–∫–∏ –≤ –∫–æ–¥–µ –∏ –∏—Ö –∏—Å–ø—Ä–∞–≤–∏—Ç—å

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def greet(self):
        return f"Hello, my name is {self.name}, and I am {self.age} years old"
# –ù–∏–∂–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–æ–∫, –µ–≥–æ –Ω–µ –Ω—É–∂–Ω–æ —É–¥–∞–ª—è—Ç—å        
bro = Person('Nikolay', 34)
assert bro.age == 34
assert bro.name == 'Nikolay'
assert bro.greet() == 'Hello, my name is Nikolay, and I am 34 years old'
sister = Person('Elena', 21)
assert sister.age == 21
assert sister.name == 'Elena'
assert sister.greet() == 'Hello, my name is Elena, and I am 21 years old'
print('Good')
–ó–∞–¥–∞—á–∞ 3 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Laptop, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –µ—Å—Ç—å: –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä init, –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π 3 –∞—Ä–≥—É–º–µ–Ω—Ç–∞: –±—Ä–µ–Ω–¥, –º–æ–¥–µ–ª—å –∏ —Ü–µ–Ω—É –Ω–æ—É—Ç–±—É–∫–∞. –ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —ç—Ç–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –Ω—É–∂–Ω–æ –¥–ª—è —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ —Å–æ–∑–¥–∞—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã brand, model, price –∏ —Ç–∞–∫–∂–µ –∞—Ç—Ä–∏–±—É—Ç laptop_name —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —Å–ª–µ–¥—É—é—â–µ–≥–æ –≤–∏–¥–∞: " " hp = Laptop('hp', '15-bw0xx', 57000) print(hp.price) # –≤—ã–≤–æ–¥–∏—Ç 57000 print(hp.laptop_name) # –≤—ã–≤–æ–¥–∏—Ç "hp 15-bw0xx" –ò –∑–∞—Ç–µ–º —Å–æ–∑–¥–∞–π—Ç–µ 2 —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –∫–ª–∞—Å—Å–∞ Laptop–∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç–µ –∏—Ö –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ laptop1 –∏ laptop2.

class Laptop:
    def __init__(self, brand, model, price):
        self.brand, self.model, self.price, self.laptop_name = brand, model, price, (brand + ' ' + model)
laptop1, laptop2 = Laptop('re', 'r', 5), Laptop('er', 'e', 47)
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–ª–∞—Å—Å–∞ Laptop –∏ –≠–ö laptop1 –∏ laptop2
assert isinstance(laptop1, Laptop)
assert isinstance(laptop2, Laptop)
hp = Laptop('hp', '15-bw0xx', 57000)
assert hp.laptop_name == 'hp 15-bw0xx'
assert hp.price == 57000
assert isinstance(hp, Laptop)
–ó–∞–¥–∞—á–∞ 4 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å SoccerPlayer, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –µ—Å—Ç—å: –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä init, –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π 2 –∞—Ä–≥—É–º–µ–Ω—Ç–∞: name, surname. –¢–∞–∫–∂–µ –≤–æ –≤—Ä–µ–º—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ–∑–¥–∞—Ç—å 2 –∞—Ç—Ä–∏–±—É—Ç–∞ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞: goals –∏ assists ‚Äî –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–æ–ª–æ–≤ –∏ –ø–µ—Ä–µ–¥–∞—á –∏–≥—Ä–æ–∫–∞, –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –æ–±–∞ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å 0. –ú–µ—Ç–æ–¥ score, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–æ–ª–æ–≤, –∑–∞–±–∏—Ç—ã—Ö –∏–≥—Ä–æ–∫–æ–º. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –¥–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ –µ–¥–∏–Ω–∏—Ü–µ. –ú–µ—Ç–æ–¥ –¥–æ–ª–∂–µ–Ω —É–≤–µ–ª–∏—á–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–±–∏—Ç—ã—Ö –≥–æ–ª–æ–≤ –∏–≥—Ä–æ–∫–∞ –Ω–∞ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. –ú–µ—Ç–æ–¥ make_assist, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ–¥–∞—á, —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –∏–≥—Ä–æ–∫–æ–º –∑–∞ –º–∞—Ç—á. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –¥–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ –µ–¥–∏–Ω–∏—Ü–µ. –ú–µ—Ç–æ–¥ –¥–æ–ª–∂–µ–Ω —É–≤–µ–ª–∏—á–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥–∞—á –∏–≥—Ä–æ–∫–æ–º –Ω–∞ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. –ú–µ—Ç–æ–¥ statistics, –∫–æ—Ç–æ—Ä—ã–π –≤—ã–≤–æ–¥ –Ω–∞ —ç–∫—Ä–∞–Ω —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∏–≥—Ä–æ–∫–∞ –≤ –≤–∏–¥–µ:‚Äã‚Äã‚Äã‚Äã‚Äã‚Äã‚Äã<–§–∞–º–∏–ª–∏—è> <–ò–º—è> - –≥–æ–ª—ã: , –ø–µ—Ä–µ–¥–∞—á–∏:

class SoccerPlayer:
    def __init__(self, name, surname):
        self.name, self.surname, self.goals, self.assists = name, surname, 0, 0
    def score(self, score = 1):
        self.goals += score
    def make_assist(self, assists = 1):
        self.assists += assists
    def statistics(self):
        print(f"{self.surname} {self.name} - –≥–æ–ª—ã: {self.goals}, –ø–µ—Ä–µ–¥–∞—á–∏: {self.assists}")
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –º–µ—Ç–æ–¥–æ–≤ –∫–ª–∞—Å—Å–∞ SoccerPlayer
leo = SoccerPlayer('Leo', 'Messi')
assert isinstance(leo, SoccerPlayer)
assert leo.__dict__ == {'name': 'Leo', 'surname': 'Messi', 'goals': 0, 'assists': 0}
leo.score(700)
assert leo.goals == 700
leo.make_assist(500)
assert leo.assists == 500
leo.statistics()
–ó–∞–¥–∞—á–∞ 5 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Zebra, –≤–Ω—É—Ç—Ä–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ –µ—Å—Ç—å –º–µ—Ç–æ–¥ which_stripe , –∫–æ—Ç–æ—Ä—ã–π –ø–æ–æ—á–µ—Ä–µ–¥–Ω–æ –ø–µ—á–∞—Ç–∞–µ—Ç —Ñ—Ä–∞–∑—ã "–ü–æ–ª–æ—Å–∫–∞ –±–µ–ª–∞—è", "–ü–æ–ª–æ—Å–∫–∞ —á–µ—Ä–Ω–∞—è", –Ω–∞—á–∏–Ω–∞—è –∏–º–µ–Ω–Ω–æ —Å —Ñ—Ä–∞–∑—ã "–ü–æ–ª–æ—Å–∫–∞ –±–µ–ª–∞—è"

class Zebra:
    def __init__(self, count = 1, color = '–ü–æ–ª–æ—Å–∫–∞ –±–µ–ª–∞—è'):
        self.count, self.color = count, color
    def which_stripe(self):
        if self.count %2 != 0:
            print(self.color)
            self.count += 1
        else:
            print('–ü–æ–ª–æ—Å–∫–∞ —á–µ—Ä–Ω–∞—è')
            self.count += 1
# –ü—Ä–æ–≤–µ—Ä–∫–∞
s = Zebra()
print(s.__dict__)
s.which_stripe()
s.which_stripe()
s.which_stripe()
‚Äã
s2 = Zebra()
s2.which_stripe()
# –í—Ç–æ—Ä–æ–µ —Ä–µ—à–µ–Ω–∏–µ —á–µ—Ä–µ–∑ –∑–∞–º–µ–Ω—É –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
class Zebra:
    def __init__(self):
        self.current = '–±–µ–ª–∞—è'
        self.next = '—á–µ—Ä–Ω–∞—è'
    
    def which_stripe(self):
        print(f'–ü–æ–ª–æ—Å–∫–∞ {self.current}')
        self.current, self.next = self.next, self.current
–ó–∞–¥–∞—á–∞ 6 –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å Person, —É –∫–æ—Ç–æ—Ä–æ–≥–æ –µ—Å—Ç—å: –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä init, –ø—Ä–∏–Ω–∏–º–∞—é—â–∏–π –∏–º—è, —Ñ–∞–º–∏–ª–∏—é –∏ –≤–æ–∑—Ä–∞—Å—Ç. –ò—Ö –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –∞—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ first_name , last_name, age. –ú–µ—Ç–æ–¥ full_name, –∫–æ—Ç–æ—Ä—ã–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –≤ –≤–∏–¥–µ "<–§–∞–º–∏–ª–∏—è> <–ò–º—è>". –ú–µ—Ç–æ–¥ is_adult, –∫–æ—Ç–æ—Ä—ã–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç True, –µ—Å–ª–∏ —á–µ–ª–æ–≤–µ–∫ –¥–æ—Å—Ç–∏–≥ 18 –ª–µ—Ç –∏ False –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

class Person:
    def __init__(self, first_name, last_name, age):
        self.first_name, self.last_name, self.age = first_name, last_name, age
    def full_name(self):
        self.full_name = self.last_name + ' ' + self.first_name
        return self.full_name
    def is_adult(self):
        if self.age >= 18:
            return True
        else:
            return False
# –ù–∏–∂–µ –∫–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –º–µ—Ç–æ–¥–æ–≤ –∫–ª–∞—Å—Å–∞ Person
p1 = Person('Ash', 'Ketchum', 20)
assert isinstance(p1, Person)
assert p1.full_name() == 'Ketchum Ash'
assert p1.age == 20
assert p1.first_name == 'Ash'
assert p1.last_name == 'Ketchum'
assert p1.is_adult() is True
```
