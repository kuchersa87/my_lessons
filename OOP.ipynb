```python
Атрибуты экземпляра класса

# У экземпляров класса также можно посмотреть их атрибуты
class Car:
    model = 'Lada'
    age = 20
a1 = Car()
a2 = Car()
# Проверим атрибуты у ЭК
a1.__dict__
# {} - пустые атрибуты
# Зададим для одного из ЭК доп атрибуты
a1.lelev = 10
print(a1.__dict__)  # атриубут появился только у того ЭК, для которого мы задали его
print(a2.__dict__)
print(Car.__dict__)  # внутри класса атрибут также не появился
# Изменили уже ранее существующий атрибут у одного из ЭК
a2.age = 100
print(a2.__dict__)   # Атрибуты внутри ЭК поменялись
print(Car.__dict__)  # Атрибуты у класса не поменялись
print(a1.__dict__)   # Ничего не поменялось
# Первым делом Питон ищет атрибуты внутри ЭК и только потом у Класса
# Удалим локальный арибут у одного из ЭК
# ВНИМАНИЕ! При этом атрибут класса остнется на месте.
del a2.age
print(a2.__dict__)
# Если вы  в 1 экземпляре класса поменяете значение атрибута, а потом удалите
# одноименный атрибут из класса,то этот атрибут останется в экземпляре класса, а во втором удалится.
Car.new = 'new'    # создали новый атрибут внутри класса
a1.new = 'not_new' # создали новый локальный атрибут внутри ЭК
print(Car.new)
del Car.new        # удалил атрибут внутри класса
print(a1.new)

Задача 1 Ниже определен пустой класс SuperHero. Ваша задача создать два ЭК и сохранить их в переменные batman и superman
class SuperHero:
    pass
batman = SuperHero()
superman = SuperHero()
setattr(batman, 'can_fly', False)
setattr(batman, 'damage', 175)
setattr(superman, 'can_fly', True)
setattr(superman, 'damage', 285)
setattr(superman, 'alter_ego', 'Кларк Кент')

Задача 2 Имеется пустой класс Config. Ваша задача написать функцию create_instance, которая принимает на вход положительное число N. Функция должна создать ЭК , создать у него N атрибутов и вернуть в качестве ответа полученный ЭК. Что касается атрибутов, они должны называться определенным образом и иметь определенное значение. В общем виде это можно записать так
class Config:     
    pass
​
def create_instance(n: int) -> Config:
    obj = Config()
    for i in range(n):
        setattr(obj, ("attribute" + str(i+1)), ("value" + str(i+1)))
    return obj
create_instance(10)
```

```python
2.4 Моносостояние для экземпляров класса Паттерн моносостояния

class Cat:
    def __init__(self, breed='', color=''):
        self.breed = breed
        self.color = color     
cat1 = Cat()
cat2 = Cat()
print(cat1.__dict__)  # Выведем словарь атрибутов cat1
print(cat2.__dict__)  # Выведем словарь атрибутов cat2
# Как видно из вывода, индивидуальные словари ЭК пока заполнены одинаковыми данными. Передадим в инициализацию параметры:
class Cat:
    def __init__(self, breed='', color=''):
        self.breed = breed
        self.color = color
cat1 = Cat('pers', 'black')
cat2 = Cat('siam', 'gray')
print(cat1.__dict__)  # Выведем словарь атрибутов cat1
print(cat2.__dict__)  # Выведем словарь атрибутов cat2
# Если вы хотите, чтобы у всех ваших экземпляров были одни общие атрибуты, вы можете воспользоваться паттерном «Моносостояние».
# Он позволяет реализовать одно состояние для атрибутов всех наших ЭК.
# Моносостояние можно создать при помощи одного общего словаря __shared_attr
class Cat:
    __shared_attr = {      # Формирование единого словаря атрибутов класса
        'breed': 'pers',   # При изменении значений - эти значения обновятся
        'color': 'black'   # во всех экземплярах класса
    }
    def __init__(self):
        self.__dict__ = Cat.__shared_attr  # Передаём в инициализатор ссылку на
                                           # созданный моно-словарь
​
cat1 = Cat()
cat2 = Cat()
print(cat1.__dict__)  # Проверим словарь атрибутов ЭК cat1
print(cat2.__dict__)  # Проверим словарь атрибутов ЭК cat
cat1.breed, cat1.color = 'pers', 'white'  # Меняем атрибуты у cat1
print('cat1:', cat1.__dict__)
print('cat2:', cat2.__dict__) # Видим, что у cat2 тоже поменялись атрибуты
cat2.breed, cat2.color = 'siam', 'gray'  # Меняем атрибуты у cat2
print('cat1:', cat1.__dict__) # Меняется и у cat1
print('cat2:', cat2.__dict__)  
# __shared_attr — общий словарь для всего пространства имён, к которому имеет доступ каждый создаваемый экземпляр класса и 
# изменения которого отражаются во всех ЭК. 
# При инициализации мы подменяем личный словарь атрибутов у ЭК на наш общий словарь __shared_attr
# Обратите внимание на то, как меняются значения в словаре: в независимости под каким ЭК были внесены новые параметры, 
# они меняются во всех экземплярах класса. 
# При этом, мы можем добавлять новые параметры через экземпляр класса, которые также будут отражены в других ЭК, например:
class Cat:
    __shared_attr = {       
        'breed': 'pers',   
        'color': 'black' 
    }
    def __init__(self):
        self.__dict__ = Cat.__shared_attr
cat1 = Cat()
cat2 = Cat()
cat1.weight = 5  # Добавляем параметр в ЭК
print('cat1:', cat1.__dict__)
print('cat2:', cat2.__dict__) 
Задача 1 https://stepik.org/lesson/682542/step/3?unit=681354

#  напишите определение класса WeatherStation    
class WeatherStation:
    __shared_attr = {"temperature": 0, "humidity": 0, "pressure": 0}
    def __init__(self):
        self.__dict__ = WeatherStation.__shared_attr
    def update_data(self, temperature, humidity, pressure):
        self.pressure, self.temperature, self.humidity = pressure, temperature, humidity
    def get_current_data(self):
        return (self.__dict__['temperature'], self.__dict__['humidity'], self.__dict__['pressure'])
        #return (self.temperature, self.humidity, self.pressure)
        #return tuple(i for i in self.__dict__.values())
        #return tuple(self.__dict__.values())
```

```python
2.5 Публичные, приватные, защищенные атрибуты и методы

# Модификаторы доступа в Python ограничивают доступ к методам класса через экземпляры класса. Их использование помогает 
# скрыть данные в классе так, чтобы извне невозможно было нарушить работу этого класса.
# Уровни доступа к атрибутам и методам класса в Python можно разделить на:
# ПУБЛИЧНЫЙ (public)
# ЗАЩИЩЁННЫЙ (protected) _
# ПРИВАТНЫЙ (private     __
# Публичный доступ
# В Python каждый атрибут класса или экземпляра класса является публичным по умолчанию. Они могут быть доступны из любой 
# точки за пределами класса.
class BankAccount:
    def __init__(self, name, balance, passport):
        self.name = name
        self.balance = balance
        self.passport = passport
    def print_data(self):   # эти данные доступны из общих вызовов
        print(self.name, self.balance, self.passport)
account1 = BankAccount('Bob', 100000, 45484564654)
account1.print_data()
print(account1.name)
print(account1.balance)
print(account1.passport)
# В итоге, мы смогли получить доступ к методу класса и к каждому атрибуту ЭК — так функционирует публичный доступ.
# Защищенный (protected) режим
# Несмотря на свое название, защищенный режим не закрывает доступ к методам и атрибутам класса. Однако существует соглашение 
# в сообществе питонистов: имя с префиксом подчеркивания, например _money, следует рассматривать как закрытую часть API (будь 
# то функция, метод или элемент данных) для внутреннего служебного использования.
# перед ним protected атрибут или метод и его не следует использовать вне класса.
class BankAccount:
    def __init__(self, name, balance, passport):
        self._name = name         # нижним подчеркиванием указываем 
        self._balance = balance   # что атрибут должен быть защищен 
        self._passport = passport # от публичного доступа
​
    def print_protected_data(self):
        print(self._name, self._balance, self._passport)
account1 = BankAccount('Bob', 100000, 45484564654)
account1.print_protected_data()
print(account1._name)      
print(account1._balance)   
print(account1._passport)
# Как видно из вывода, данные всё так же доступны для использования вне класса.
# Приватный (private) режим
# Приватный режим доступа (private), который ограничивает доступ вне класса.
# Давайте попробуем получить доступ к защищенным атрибутам через специально разрешенный метод:
class BankAccount:
    def __init__(self, name, balance, passport):
        self.__name = name
        self.__balance = balance
        self.__passport = passport
    def print_private_data(self):
        return self.__name, self.__balance, self.__passport
account1 = BankAccount('Bob', 100000, 45484564654)
print(account1.print_private_data())
# В результате мы можем получать доступ к данным через специально организованный метод print_private_data
# Но доступ к атрибутам здесь осуществляется внутри самого класса.
print(account1.__name) # AttributeError
# класс BankAccount не имеет атрибута __name, тем самым показывая нам сокрытие атрибутов класса от внешнего доступа.
# Такое сокрытие обработки защищенных атрибутов называется инкапсуляция: мы ограждаем всех, кто пользуется нашим классом 
# BankAccount от доступа к защищенным атрибутам, но внутри самого класса доступ к ним неограничен.
           # Выведет нам необходимые данные
# Аналогичная ситуация и с приватным методом: вне класса нельзя обратиться к его имени, внутри класса мы можем это сделать. 
# В примере ниже мы вызываем приватный метод __print_private_data внутри открытого метода public_call
class BankAccount:
    def __init__(self, name, balance, passport):
      self.__name = name
      self.__balance = balance
      self.__passport = passport
    def public_call(self):
      print('work public method')      
      self.__print_private_data()  # Обращение к приватному методу
    def __print_private_data(self):
      print('work private method')
      print(self.__name, self.__balance, self.__passport)
account1 = BankAccount('Bob', 100000, 45484564654)
account1.public_call()           # Выведет нам необходимые данные
account1.__print_private_data()  # AttributeError: 'BankAccount' object has no attribute '__print_private_data'
# Как получить доступ к приватным атрибутам и методам вне класса
# В заключение необходимо отметить, что приватность доступа к атрибутам и методам класса — это лишь общее соглашение вроде 
# "в нашем клубе принято верить джентльменам на слово".
# Для этого мы можем узнать список имен, доступных в нашем классе,  при помощи функции dir:
class BankAccount:
    def __init__(self, name, balance, passport):
        self.__name = name
        self.__balance = balance
        self.__passport = passport
    def public_call(self):
        self.__print_private_data()
    def __print_private_data(self):
        print(self.__name, self.__balance, self.__passport)
account1 = BankAccount('Bob', 100000, 45484564654)
print(dir(account1))  # просмотр атрибутов нашего экземпляра класса
на приватные данные
# Ещё один метод обратиться к ключам, по которым можно выйти на приватные данные
print(PizzaMaker.__dict__.keys())
# Обратите внимание на такие названия, как:
# _BankAccount__balance
# _BankAccount__name
# _BankAccount__passport
# _BankAccount__print_private_data
# Именно через них мы сможем получить доступ к приватным данным вне класса:
class BankAccount:
    def __init__(self, name, balance, passport):
        self.__name = name
        self.__balance = balance
        self.__passport = passport
    def public_call(self):
        self.__print_private_data()
    def __print_private_data(self):
        print(self.__name, self.__balance, self.__passport)
account1 = BankAccount('Bob', 100000, 45484564654)
print(account1._BankAccount__balance)
print(account1._BankAccount__name)
print(account1._BankAccount__passport)
account1._BankAccount__print_private_data()
Задача 1 https://stepik.org/lesson/682543/step/9?unit=681355

class AverageCalculator:
    def __init__(self, numbers):
        self.numbers = numbers
​
    def __calculate_average(self):
        total = sum(self.numbers)
        return total / len(self.numbers)
average_calculator = AverageCalculator([1, 2, 3])
# print(dir(average_calculator))  # просмотр атрибутов нашего экземпляра класса
print(average_calculator._AverageCalculator__calculate_average())
Задача 2 https://stepik.org/lesson/682543/step/10?unit=681355

class PizzaMaker:
    def __make_pepperoni(self):
        pass
    def _make_barbecue(self):
        pass
proba = PizzaMaker()
print(PizzaMaker.__dict__.keys())
print('*'*30)
print(dir(proba))
maker = PizzaMaker()
maker._make_barbecue()
maker._PizzaMaker__make_pepperoni()
Задача 3
https://stepik.org/lesson/682543/step/11?unit=681355
class Student:
    def __init__(self, name, age, branch):
        self.__name, self.__age, self.__branch = name, age, branch
    def __display_details(self):
        print(f'Имя: {self.__name}', f'Возраст: {self.__age}', f'Направление: {self.__branch}', sep = '\n')
    def access_private_method(self):
        self.__display_details()
Задача 4
https://stepik.org/lesson/682543/step/12?unit=681355
class BankDeposit:
    def __init__(self, name, balance, rate):
        self.name, self.balance, self.rate = name, balance, rate
    def __calculate_profit(self):
        return self.balance * self.rate * 0.01
    def get_balance_with_profit(self):
        return self.__calculate_profit() + self.balance
Задача 5
https://stepik.org/lesson/682543/step/13?unit=681355
class Library:
    def __init__(self, books):
        self.__books = books
    def __check_availability(self, book_to_check):
        if book_to_check in self.__books:
            return True
        else:
            return False
    def search_book(self, find_book):
        return self.__check_availability(find_book)
    def return_book(self, book_add_to_list):
        self.__books.append(book_add_to_list)
    def _checkout_book(self, book):
        if book in self.__books:
            self.__books.remove(book)
            return True
        else:
            return False
Задача 6
https://stepik.org/lesson/682543/step/14?unit=681355
class Employee:
    def __init__(self, name, position, hours_worked, hourly_rate):
        self.name, self.__position, self.__hours_worked, self.__hourly_rate = name, position, hours_worked, hourly_rate
    def __calculate_salary(self):
        return self.__hours_worked * self.__hourly_rate
    def _set_position(self, position_corrected):
        self.__position = position_corrected
    def get_position(self):
        return self.__position
    def get_salary(self):
        return self.__calculate_salary()
    def get_employee_details(self):
        return f"Name: {self.name}, Position: {self.__position}, Salary: {self.__calculate_salary()}"
```
