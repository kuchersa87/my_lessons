## Пропуски / coalesce

```python
# В таком случае count пропускает пропуски - не считает их - count(t.checking_account)
# В таком случае count считает каждую строку - count(1)
sql = '''select count(t.checking_account), count(1) from german_credit t'''
```

```python
select(sql)
```

```python
# Так по умолчанию пропуски тоже будут учитываться!
sql = '''
select 
t.checking_account,

count(*) as cnt,
avg(t.credit_amount) as credit_amount_avg

 from german_credit t
 group by t.checking_account
 '''
```

```python
select(sql)
```

```python
# В pandas по умолчанию пропускт не считает
df.groupby('checking_account')['credit_amount'].count()
```

```python
# Необходимо явно задать считалку по пропускам
df.groupby('checking_account',dropna=False)['credit_amount'].count()
```

```python
# Флаг по пропускам
# is null - пропуск
sql = '''select t.checking_account,
case when t.checking_account is null then 1 else 0 end as is_null
from german_credit t'''
```

```python
import csv
from io import StringIO

def psql_insert_copy(table, conn, keys, data_iter):
    # gets a DBAPI connection that can provide a cursor
    dbapi_conn = conn.connection
    with dbapi_conn.cursor() as cur:
        s_buf = StringIO()
        writer = csv.writer(s_buf)
        writer.writerows(data_iter)
        s_buf.seek(0)

        columns = ', '.join('"{}"'.format(k) for k in keys)
        if table.schema:
            table_name = '{}.{}'.format(table.schema, table.name)
        else:
            table_name = table.name

        sql = 'COPY {} ({}) FROM STDIN WITH CSV'.format(
            table_name, columns)
        cur.copy_expert(sql=sql, file=s_buf)
```

```python
users.to_sql('users',con_pg,index=False,if_exists='replace',method=psql_insert_copy)
items.to_sql('items',con_pg,index=False,if_exists='replace',method=psql_insert_copy)
```
